## Q1) Looking up for Flights based on user entered Source, Destination and Class

```python
# Ask for user for input
source = input("Enter Source City: ")
destination = input("Enter Destination City: ")
travel_class = input("Enter Class (e.g. Business/Economy): ")

# Filter dataset
filtered = data[
    (data['source_city'] == source) &
    (data['destination_city'] == destination) &
    (data['class'] == travel_class)
]

# Group by airline and calculate average price, sorted ascending
if not filtered.empty:
    result = (
        filtered.groupby('airline')['price']
        .mean()
        .reset_index()
        .sort_values(by='price', ascending=True)
    )
    print(f"\nAvailable flights from {source} to {destination} in {travel_class} class :\n")
    print(result.rename(columns={'Airline': 'Airline', 'Price': 'Average Price'}))
else:
    print("\nNo matching flights found for your input.")
```

## Q2) What are the airlines in the dataset (with their frequency)?

```python
#extracting names of airlines and their frequency of flying + bar graph
data['airline'].unique()
data['airline'].value_counts()
```

```python
data['airline'].value_counts(ascending=True).plot.barh(color = 'orange')
plt.title("Airline with Frequency")
plt.xlabel("Number of Flights")
plt.ylabel("Airline Names")
plt.show()
```

## Q3) Are non-stop flights significantly more expensive than flights with 1+ stops?

```python
order_stops = ['zero', 'one', 'two_or_more']
stops_price = data.groupby(['stops','airline'])['price'].mean().reset_index()
plt.figure(figsize=(12,6))
sns.barplot(x='stops', y='price', hue='airline',data=stops_price, order=order_stops,errorbar=None, palette="Set2")
plt.title("Average Flight Price: Non-stop vs Stops (by Airline)")
plt.xlabel("Number of Stops")
plt.ylabel("Average Price")
plt.legend(title="Airline")
plt.show()
```

## Q4) Does price varies with airlines ?

```python
#categorical plot
sns.catplot(x = 'airline', y = 'price', hue='airline', kind ='bar', palette="rocket", data = data,legend=True)
plt.title("Price variation among different Airline ")
plt.xlabel("Airlines")
plt.ylabel("Avg Price")
plt.show()
```

## Q5) Does ticket price change based on the departure time?

```python
# checking the mean ticket price by departure time
data.groupby(['departure_time','class'])['price'].mean().round(2)
```

```python
# User input for source and destination
source = input("Enter Source City: ")
destination = input("Enter Destination City: ")

# Filter dataset for the chosen route
route = data[(data['source_city'] == source) & (data['destination_city'] == destination)]

# Check if flights exist for that route
if route.empty:
    print(f"No flights found from {src} to {dest}.")
else:
    plt.figure(figsize=(10,6))
    sns.pointplot(
        x='departure_time', y='price',
        hue='class', data=route,
        order=['Early_Morning','Morning','Afternoon','Evening','Night','Late_Night'],
        palette="Dark2", errorbar=None
    )
    plt.title(f"Average Price by Departure Time and Class ({source} → {destination})")
    plt.ylabel("Average Price")
    plt.xlabel("Departure Time")
    plt.show()
```

## Q6) How is the price affected when tickets are bought 1 or 2 days before departure?

```python
grouped = data.groupby(['days_left','class'])['price'].mean().reset_index()
grouped
```

```python
plt.figure(figsize=(12,6))
sns.pointplot(x='days_left', y='price', hue='class', data=grouped , legend= True)
plt.title("Average Price by Days Left")
plt.xlabel("Days Left")
plt.ylabel("Average Price")
plt.show()
```
